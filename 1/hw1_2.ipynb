```python
import pandas as pd
import matplotlib.pyplot as plt
rawData = pd.read_csv('HW1_data.csv')

# function: find average, and append
def findAvg(list, listLen):
    cnt = listLen
    sum = 0
    for i in range(listLen):
        if list[i] != 0:
            sum += int(list[i])
        else:
            cnt -= 1
    list.append( str(round(sum/cnt, 2)) )

# function: find max, and return index
def findMax(list, listLen):
    max = 0
    maxIndex = 0
    for i in range(listLen):
        if int(list[i]) > max:
            max = int(list[i])
            maxIndex = i
    return maxIndex

# record all names
name = []
for i in range(len(rawData)):
    if(not (rawData.iloc[i,0] in name)):
        name.append(rawData.iloc[i, 0])

# reconstruct dataframe
listLen = len(name)
sex = ['0'] * listLen
age = ['0'] * listLen
hr = ['0'] * listLen
height = ['0'] * listLen
weight = ['0'] * listLen
bp = ['0'] * listLen
for i in range(len(rawData)):
    index = name.index(rawData.iloc[i, 0])
    eval( rawData.iloc[i, 1].lower() )[index] = rawData.iloc[i, 2]

# append average row
name.append('Average')
sex.append('0')
findAvg(age, listLen)
findAvg(hr, listLen)
findAvg(height, listLen)
findAvg(weight, listLen)
findAvg(bp, listLen)

# show dataframe
dict = {'Name': name, 'Sex': sex, 'Age': age, 'HR': hr, 'Height': height, 'Weight': weight, 'BP': bp}
df = pd.DataFrame(dict)
print(df)

print('--------------------------------------------------------------')
print('Max age: %s' %( name[findMax(age, listLen)] ))
print('Max HR: %s' %( name[findMax(hr, listLen)] ))
print('Max height: %s' %( name[findMax(height, listLen)] ))
print('Max weight: %s' %( name[findMax(weight, listLen)] ))
print('Max BP: %s' %( name[findMax(bp, listLen)] ))

# scatter plot of height and weight
plt.subplot(2, 2, 1)    # upper left
plt.title("Scatter plot")
maleHeight = []
maleWeight = []
femaleHeight = []
femaleWeight = []
for i in range(listLen):
    if sex[i] == 'M' and int(height[i]) and int(weight[i]):
        maleHeight.append(int(height[i]))
        maleWeight.append(int(weight[i]))
    elif sex[i] == 'F' and int(height[i]) and int(weight[i]):
        femaleHeight.append(int(height[i]))
        femaleWeight.append(int(weight[i]))
plt.scatter(maleHeight, maleWeight, c ="blue")
plt.scatter(femaleHeight, femaleWeight, c ="red")
plt.xlabel("Height")
plt.ylabel("Weight")

# pie chart of age
plt.subplot(2, 2, 2)    # upper right
plt.title("Pie chart of age")
sectionCnt = [0] * 8
label = ['1-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80']
del age[-1] # delete last element (avg)
for i in range(listLen):
    section = int( (int(age[i])-1)/10 )
    sectionCnt[section] += 1
plt.pie(sectionCnt, labels = label)
plt.show()

# histogram of age
plt.subplot(2, 2, 3)    # lower left
plt.title("Histogram")
plt.xlabel('age')
plt.ylabel('count')
age = [int(i) for i in age]
plt.hist(age)

# pie chart of Sex
plt.subplot(2, 2, 4)    # lower right
plt.title("Pie chart of sex")
sexCnt = [0] * 2
for i in range(listLen):
    if sex[i] == 'M':
        sexCnt[0] += 1
    elif sex[i] == 'F':
        sexCnt[1] += 1
plt.pie(sexCnt, labels = ['Male', 'Female'])
```
