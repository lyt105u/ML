```python
import pandas as pd
import numpy as np
rawData = pd.read_csv('HW1_data.csv')

# function: find average, and append
def findAvg(list, listLen):
    cnt = listLen
    sum = 0
    for i in range(listLen):
        # if list[i].isdigit():
        if not pd.isnull(list[i]):
            sum += int(list[i])
        else:
            cnt -= 1
    list.append( str(round(sum/cnt, 2)) )

# record all names
name = []
for i in range(len(rawData)):
    if(not (rawData.iloc[i,0] in name)):
        name.append(rawData.iloc[i, 0])

# reconstruct dataframe
listLen = len(name)
sex = [np.NaN] * listLen
age = [np.NaN] * listLen
hr = [np.NaN] * listLen
height = [np.NaN] * listLen
weight = [np.NaN] * listLen
bp = [np.NaN] * listLen
for i in range(len(rawData)):
    index = name.index(rawData.iloc[i, 0])
    eval( rawData.iloc[i, 1].lower() )[index] = rawData.iloc[i, 2]

# append average row
name.append('Average')
sex.append(np.NaN)
findAvg(age, listLen)
findAvg(hr, listLen)
findAvg(height, listLen)
findAvg(weight, listLen)
findAvg(bp, listLen)

# show dataframe
dict = {'Name': name, 'Sex': sex, 'Age': age, 'HR': hr, 'Height': height, 'Weight': weight, 'BP': bp}
df = pd.DataFrame(dict)
print(df)

print('--------------------------------------------------------------')
print('Max age: %s' %( name[age.index(max(age))] ))
print('Max HR: %s' %( name[hr.index(max(hr))] ))
print('Max height: %s' %( name[height.index(max(height))] ))
print('Max weight: %s' %( name[weight.index(max(weight))] ))
print('Max BP: %s' %( name[bp.index(max(bp))] ))
```
